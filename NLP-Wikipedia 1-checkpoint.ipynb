### Textvorverarbeitung (Preprocessing)

Es wird eine Funktion namens clean_text definiert, welche die Textvorverarbeitung durchführt. Dabei sollen drei Schritte ausgeführt werden: die Normalisierung der Groß- und Kleinschreibung, das Entfernen von Satzzeichen und das Eliminieren von numerischen Ausdrücken. Diese Vorverarbeitungsschritte dienen dazu, den Text für weitere Analysezwecke vorzubereiten.

```python
def clean_text(text):
    text = text.str.lower()
    text = text.str.replace("[^\w\s]", "")
    text = text.str.replace("\d", "")
    text = text.str.replace("\n", "")
    return text
```

```python
df["text"] = clean_text(df["text"])
df["text"].head()
```

### Entfernen von Stoppwörtern

Es wird eine Funktion namens remove_stopwords erstellt. Sie soll dabei helfen, Schlüsselmerkmale aus dem Text zu extrahieren, indem sie Wörter entfernt, die als unwichtig oder irrelevant betrachtet werden.

```python
import nltk
nltk.download("stopwords")
```

```python
def remove_stopwords(text):
    sw = stopwords.words("english")
    text = text.apply(lambda x: " ".join(x for x in str(x).split() if x not in sw))
    return text
```

```python
df["text"] = remove_stopwords(df["text"])
df["text"].head()
```

 Alle Schritte werden nun in einer einzigen Funktion zusammengefasst, die zusätzlich die Möglichkeit bietet, Barplot- und WordCloud-Visualisierungen zu generieren.

```python
def text_preprocess (text, Barplot=False, Wordcloud=False):
    text = text.str.lower()
    text = text.str.replace("[^\w\s]", "")
    text = text.str.replace("\d", "")
    text = text.str.replace("\n", '')
    # import nltk
    # nltk.download("stopwords")
    sw = stopwords.words("english")
    text = text.apply(lambda x: " ".join(x for x in str(x).split() if x not in sw))
    temp_df = pd.Series(" ".join(text).split()).value_counts()
    drops = temp_df[temp_df < 1000]
    text = text.apply(lambda x: " ".join(x for x in x.split() if x not in drops))
    # nltk.download("wordnet")
    text = text.apply(lambda x: " ".join([Word(word).lemmatize() for word in x.split()]))


    if Barplot:
        tf = text.apply(lambda x: pd.value_counts(x.split(" "))).sum(axis=0).reset_index()
        tf.columns = ["words", "tf"]
        tf[tf["tf"] > 5000].plot.bar(x="words", y="tf")
        plt.show(block=True)

    if Wordcloud:
        text = " ".join(i for i in text)
        worldcloud = WordCloud().generate(text)
        plt.imshow(worldcloud, interpolation="bilinear")
        plt.axis("off")
        plt.show(block=True)

    return text

df["text"] = text_preprocess(df["text"], Barplot=True, Wordcloud=True)
```

Und nun noch ein Docstring, der die Funktion erklärt:

```python
def text_preprocess(text, Barplot=False, Wordcloud=False):
    """
    Preprocessing the provided texts.

    Args:
        text: variable with texts in DataFrame
        Barplot: Barplot visualisation, optional
        Wordcloud: Wordcloud vusualisation, optional

    Returns:
        text

    Example:
        text_preprocess(dataframe[column_name], Barplot=True, Wordcloud=True)

    """


    # Normalizing Case Folding
    text = text.str.lower()
    # Punctuations
    text = text.str.replace("[^\w\s]", "")
    text = text.str.replace("\n", '')
    # Numbers
    text = text.str.replace("\d", "")
    # Stopwords
    # import nltk
    # nltk.download("stopwords")
    sw = stopwords.words("english")
    text = text.apply(lambda x: " ".join(x for x in str(x).split() if x not in sw))
    # Rarewords
    temp_df = pd.Series(" ".join(text).split()).value_counts()
    drops = temp_df[temp_df < 1000]
    text = text.apply(lambda x: " ".join(x for x in x.split() if x not in drops))
    # Lemmatization
    # nltk.download("wordnet")
    text = text.apply(lambda x: " ".join([Word(word).lemmatize() for word in x.split()]))

    if Barplot:
        # Calculation of Term Frequencies
        tf = text.apply(lambda x: pd.value_counts(x.split(" "))).sum(axis=0).reset_index()
        # Naming of the Columns
        tf.columns = ["words", "tf"]
        # Visualization of words occurring more than 5000 times
        tf[tf["tf"] > 5000].plot.bar(x="words", y="tf")
        plt.show(block=True)

    if Wordcloud:
        # Combining Words
        text = " ".join(i for i in text)
        # Specifying the Features of Word Cloud Visualization
        wordcloud = WordCloud(max_font_size=50,
                              max_words=100,
                              background_color="white").generate(text)
        plt.figure()
        plt.imshow(worldcloud, interpolation="bilinear")
        plt.axis("off")
        plt.show(block=True)

    return text
```

```python
df["text"] = text_preprocess(df["text"], Barplot=True, Wordcloud=True)
```
